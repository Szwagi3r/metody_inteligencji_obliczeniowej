# Prostokątna sieć Kohonena

```python
import pandas as pd
import numpy as np
import seaborn as sns
import math
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
class SOM:
    
    def __init__(self, M, N, radius=1, data_dim=2):
        self.M = M
        self.N = N
        self.radius = radius
        self.neurons = np.random.normal(0, 3, size=(N*M, data_dim))
        
    @staticmethod
    def net_distance(self, x,y):
        """
        odległość między neuronami x, y na siatce dyskretnej
        odległość jest liczona za pomocą tw. Pitagorasa
        """
        return self.radius * math.sqrt((x%(self.N)-y%(self.N))**2 + (math.floor(x/self.M)-math.floor(y/self.M))**2)
    
    @staticmethod
    def decay(t,lam):
        """
        funckja alfa wygaszająca uczenie
        """
        return np.exp(-t/lam)
    
    @staticmethod
    def gauss(x, t):
        """
        funkcja gaussowska dla sąsiedztwa
        """
        return np.exp(-(x*t)**2)
    
    @staticmethod
    def mexican_hat(x, t):
        """
        funkcja meksykański kapelusz dla sąsiedztwa
        """
        return (2-4*(x*t)**2) * t**2 * np.exp(-(x*t)**2)
    
    def train(self, data, epochs=100, neighborhood="gauss"):
        """
        trenowanie sieci
        """
        # wybór funkcji sąsiedztwa
        if neighborhood == "gauss":
            neighborhood_fun = self.gauss
        elif neighborhood == "mexican_hat":
            neighborhood_fun = self.mexican_hat
        else:
            print("Unknown neighborhood function, assuming gauss")
            neighborhood_fun = self.gauss
        
        # trenowanie 
        for epoch in range(epochs):
            print("Epoch: {}/{}".format(epoch+1,epochs), end='\r')
            obs_permutation = np.random.permutation(len(data))
            for obs in data[obs_permutation]:
                # znajdowanie zwycięzkiego neuronu dla obserwacji
                winner = np.argmin([np.linalg.norm(obs-neuron) for neuron in self.neurons])
                # przesunięcie wszystkich neuronów
                for i, neuron in enumerate(self.neurons):
                    self.neurons[i] += neighborhood_fun(self.net_distance(self, winner, i), epoch)*self.decay(epoch, epochs)*(obs-neuron)
    
    def predict(self, data):
        """
        dla każdej obserwacji znajduje najbliższy neuron 
        """
        return [np.argmin([np.linalg.norm(neuron-obs) for neuron in self.neurons]) for obs in data]
```

## Zbiory danych 

```python
df1 = pd.read_csv('datasets/mio2/hexagon.csv')
x1 = np.asarray(df1.iloc[:,0:2])
df1
```

```python
df2 = pd.read_csv('datasets/mio2/cube.csv')
x2 = np.asarray(df2.iloc[:,0:3])
df2
```

```python
plt.figure(figsize=(6, 6), dpi=60)
plt.scatter(x=df1['x'], y=df1['y'], c=df1['c'], s=15)
plt.title('Hexagon', fontsize=20)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df2['x'], df2['y'], df2['z'], c=df2['c'])
plt.title('Cube', fontsize=20)
plt.show()
```

### Funkcja gaussowska jako funkcja sąsiedztwa

#### Zbiór hexagon

```python
som = SOM(M=2, N=3, radius=1, data_dim=2)
som.train(data=x1, epochs=100)
pred = som.predict(x1)

plt.scatter(x=df1['x'], y=df1['y'], c=pred, s=10)
plt.scatter(som.neurons[:,0], som.neurons[:,1], color = "black")
plt.title('Hexagon', fontsize=20)
plt.show()
```

Wprzypadku określenia z góry 6 klastrów, dla wartości $r=1$ dostaliśmy bardzo dobre klastrowanie (i podobne do tego z oryginalnych etykiet). Sprawdźmy teraz, jak sieć poradzi sobie, w przypadku gdy neuronów będzie więcej.

```python
som = SOM(M=3, N=4, radius=1, data_dim=2)
som.train(data=x1, epochs=100)
pred = som.predict(x1)

plt.scatter(x=df1['x'], y=df1['y'], c=pred, s=10)
plt.scatter(som.neurons[:,0], som.neurons[:,1], color = "black")
plt.title('Hexagon', fontsize=20)
plt.show()
```

Dla siatki $3\times 4$ neuronów, niektóre klastry podzieliły się na mniejsze. Gdybyśmy połączyli ze sobą klastry o najbliższych neuronach, podział byłby podobny do oczekiwanego.

#### Zbiór cube

```python
som = SOM(M=2, N=4, radius=1, data_dim=3)
som.train(data=x2, epochs=100)
pred = som.predict(x2)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df2['x'], df2['y'], df2['z'], c=pred)
ax.scatter(som.neurons[:,0], som.neurons[:,1], som.neurons[:,2], color = "black", s=30)
plt.title('Cube', fontsize=20)
plt.show()
```

Mimo że wykres jest średnio czytelny (jak to 3D) możemy zauważyć że klastry układają się w wierzchołkach sześcianu, więc w przypadku danych trójwymiarowych, redukcja wymiarów na siatkę prostokątną zdała egzamin. Sprawdźmy co się stanie dla większej liczby klastrów (neuronów)

```python
som = SOM(M=2, N=5, radius=1, data_dim=3)
som.train(data=x2, epochs=100)
pred = som.predict(x2)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df2['x'], df2['y'], df2['z'], c=pred)
ax.scatter(som.neurons[:,0], som.neurons[:,1], som.neurons[:,2], color = "black", s=30)
plt.title('Cube', fontsize=20)
plt.show()
```

Dla 10 klastrów i więcej ciężko już jest odróżnić gdzie znajdują się konkretne klastry, jednak podział wygląda na zadowalająco dobry.

### Funkcja meksykański kapelusz jako funkcja sąsiedztwa

#### Zbiór hexagon

```python
som = SOM(M=2, N=3, radius=1, data_dim=2)
som.train(data=x1, epochs=100, neighborhood='mexican_hat')
pred = som.predict(x1)

plt.scatter(x=df1['x'], y=df1['y'], c=pred, s=10)
plt.scatter(som.neurons[:,0], som.neurons[:,1], color = "black")
plt.title('Hexagon', fontsize=20)
plt.show()
```

Ze względu na odpychanie spowodowne ujemnymi wartościami meksykańskiego kapelusza, w 4 iteracji neurony rozbiegły do zbyt dużych wartości. Działo się tak zarówno dla bardzo małych jak i dużych wartości r.

#### Zbiór cube

```python
som = SOM(M=2, N=4, radius=0.4, data_dim=3)
som.train(data=x2, epochs=100, neighborhood='mexican_hat')
pred = som.predict(x2)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df2['x'], df2['y'], df2['z'], c=pred)
ax.scatter(som.neurons[:,0], som.neurons[:,1], som.neurons[:,2], color = "black", s=30)
plt.title('Cube', fontsize=20)
plt.show()
```

Tutaj sytuacja analogiczna, w żadnym przypadku nie udało się zmusić kapelusza do stablizacji.

## Wnioski
+ funkcja Gaussa jako funkcja sąsiedztwa poradziła sobie dobrze, klastrowanie jest poprawne i pokrywa się z oczekiwanym podziałem.
+ dla danych zbiórów nie udało się nauczyć sieci używając meksykańskiego kapelusza jako funkcji sąsiedztwa, dla różnych wartości parametru $r$ (również spoza przedziału $[0.1, 1]$). Jednym z możliwych rozwiązań problemu byłoby dobranie funkcji o mniejszych, bądź bradziej płaskich fragmentów, gdzie wartości fukncji są ujemne.
